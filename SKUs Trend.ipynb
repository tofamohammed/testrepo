```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales2.csv')
```

```python
df.head()
```

```python
df.drop(["Unnamed: 13"], axis=1, inplace=True)
df.head()
```

```python
df.rename(columns = {'Unnamed: 0':'Product'}, inplace = True)
df.head()
```

```python
df.set_index('Product', inplace=True)
df.head()
```

```python
year = list(map(str, range(1, 13)))
year
```

```python
df.index.name = None
df.head()
```

```python
df.loc['2018EDIBLE'].dtypes
```

```python
df.head()
```

```python
# Assuming 'df' is your DataFrame
df = df.replace(',', '', regex=True)  # Remove commas from the data
df = df.astype(int)  # Convert DataFrame to integer

# Print the DataFrame to verify
print(df)
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
edible_18 = df.loc['2018EDIBLE', months] # passing in years 1980 - 2013 to exclude the 'total' column
edible_18.head()
```

```python
edible_18.plot()
```

```python
edible_18.plot(kind='line')

plt.title('2018 Edible Trend')
plt.ylabel('Metric Tons')
plt.xlabel('Months')
```

```python
edible_18.plot(kind='line')

plt.title('2018 Edible Trend')
plt.ylabel('Metric Tons')
plt.xlabel('Months')
#plt.text(6, 20000, '2018 Big Sallah') # see note below

plt.show()
```

```python
edible_1819 = df.loc[['2018EDIBLE','2019EDIBLE'], months]
edible_1819
```

```python
edible_1819.plot()
```

```python
edible_1819 = edible_1819.transpose()
edible_1819.head()
```

```python
edible_1819.plot(kind='line')

plt.title('Edible Trend 2018 and 2019')
plt.ylabel('Metric Tons')
plt.xlabel('Months')

plt.show()
```
